# Statistical Regularity

```python
import random
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python

random.seed(44)
trials =[]
outcomes =[]
relative_freq = []
h=0;t=0
for x in range(1,1001):
    trials.append(x)
    outcomes.append(random.choice(['H','T']))
    if outcomes[-1] == 'H':
        h += 1
        relative_freq.append(h/trials[-1])
    else:
        t += 1
        relative_freq.append(t/trials[-1])

df = pd.DataFrame({'trial_No' : trials, 'outcomes' : outcomes, 'relative_freq' : relative_freq})

plt.scatter()
```
