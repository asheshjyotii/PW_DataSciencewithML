# Handling Outliers

## Finding the Five point summary

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
salary =[
    45000, 48000, 50000, 52000, 53000, 55000, 56000, 57000, 58000, 59000,
    60000, 61000, 62000, 63000, 64000, 65000, 66000, 67000, 68000, 69000,
    70000, 71000, 72000, 73000, 74000, 75000, 76000, 77000,
    20000,110000
]


df_quantiles = pd.DataFrame(np.quantile(salary,[0,0.25,0.50,0.75,1]), columns = ['Salary'])
df = pd.DataFrame(salary, columns = ['Salary'])

df_quantiles.head()
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1) # number of rows, number of columns, index
sns.histplot(df['Salary'],kde = True)
plt.title("Distribution Plot")
plt.subplot(1,2,2)
sns.boxplot(data=df, x = 'Salary')
plt.title("Box Plot")
plt.show()
```

```python
Q1 = df['Salary'].quantile(0.25)
Q3 = df['Salary'].quantile(0.75)

IQR = Q3-Q1

lower_fence = Q1 - 1.5*IQR
upper_fence = Q3 + 1.5*IQR

df_filtered = df[(df.Salary>=lower_fence) & (df.Salary<=upper_fence)]
```

```python
plt.figure(figsize = (12,5))
plt.subplot(1,2,1)
plt.grid(axis = 'both')
sns.histplot(df_filtered, kde=True)
plt.title("Distribution Plot")


plt.subplot(1,2,2)
sns.boxplot(df_filtered, x = 'Salary')
plt.title("Box Plot")
plt.grid(axis = 'both')
plt.show()
```

# Imputing outilers with mean

1. 

```python
Q1 = df["Salary"].quantile(0.25)
Q3 = df["Salary"].quantile(0.75)

IQR = Q3-Q1

lower_fence = Q1-1.5*IQR
upper_fence = Q3+1.5*IQR

mean = df[(df["Salary"]>=lower_fence) & (df["Salary"]<=upper_fence)].mean()
```

```python
df["Salary"] = np.where((df["Salary"]<lower_fence) | (df["Salary"]>upper_fence), mean, df["Salary"])
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.grid(True)
sns.histplot(df["Salary"],kde=True)

plt.subplot(1,2,2)
plt.grid(True)
sns.boxplot(data = df, x = "Salary")
plt.show()
```
